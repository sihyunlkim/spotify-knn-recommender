```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.neighbors import NearestNeighbors

# ==========================================
# 1. Data Loading & Preprocessing
# ==========================================
def load_and_preprocess_data(filepath):
    #  
    df = pd.read_csv(filepath)
    print(f"Original Data Size: {df.shape}")

    # 1-1. Data Cleaning (Handling Anomalies)
    # valence 0      (tempo=0   )
    df_clean = df[df['valence'] > 0].copy()
    
    # 1-2. Drop Duplicates
    #  ID    
    df_clean = df_clean.drop_duplicates(subset=['id'])
    #        (  )
    df_clean = df_clean.drop_duplicates(subset=['name', 'artists'])
    
    print(f"Cleaned Data Size: {df_clean.shape}")
    
    return df_clean

#   
df_final = load_and_preprocess_data('data/data.csv')

# ==========================================
# 2. Feature Engineering & Scaling
# ==========================================
#     
feature_cols = [
    'acousticness', 'danceability', 'energy', 'instrumentalness',
    'liveness', 'loudness', 'speechiness', 'tempo', 'valence',
    'popularity', 'year'
]

# Standard Scaling ( 0,  1 )
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_final[feature_cols])

#    DataFrame  (  !)
df_scaled = pd.DataFrame(X_scaled, columns=feature_cols, index=df_final.index)

# ==========================================
# 3. Clustering (K-Means) & PCA Visualization
# ==========================================
# 3-1. K-Means Clustering
# Elbow Method  , Silhouette Score  k=5  
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(df_scaled)

# 3-2. PCA (Dimensionality Reduction)
# 11 2  
pca = PCA(n_components=2)
pca_coords = pca.fit_transform(df_scaled)

#   
pca_df = pd.DataFrame(pca_coords, columns=['PC1', 'PC2'], index=df_final.index)
pca_df['Cluster'] = clusters
pca_df['name'] = df_final['name'] #  

# 3-3. Global Visualization (Deliverable 1)
plt.figure(figsize=(12, 8))
sns.scatterplot(
    x='PC1', y='PC2', hue='Cluster', data=pca_df, 
    palette='viridis', alpha=0.6, s=15
)
plt.title('Music Feature Space: PCA with K-Means Clustering (k=5)', fontsize=15)
plt.xlabel('PC1 (Intensity & Energy)')
plt.ylabel('PC2 (Atmosphere & Acoustics)')
plt.legend(title='Genre/Mood Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()

# ==========================================
# 4. Recommendation Model (kNN)
# ==========================================
#       
# n_neighbors=6 (  1 +   5)
knn_model = NearestNeighbors(n_neighbors=6, metric='euclidean')
knn_model.fit(df_scaled)

def recommend_song(song_name, artist_name=None):
    """
         5   
    """
    # --- Step 1:   ---
    try:
        #    ( )
        mask = df_final['name'].str.lower() == song_name.lower()
        
        #      
        if artist_name:
            mask = mask & df_final['artists'].str.lower().str.contains(artist_name.lower())
        
        found_songs = df_final[mask]
        
        if len(found_songs) == 0:
            print(f"Error:  '{song_name}' (Artist: {artist_name})   .")
            return

        #      
        target_idx = found_songs.index[0]
        print(f"Selected Song: {df_final.loc[target_idx, 'name']} - {df_final.loc[target_idx, 'artists']}")

    except Exception as e:
        print(f"Error during search: {e}")
        return

    # --- Step 2:    (kNN Query) ---
    input_features = df_scaled.loc[[target_idx]]
    distances, indices = knn_model.kneighbors(input_features)
    
    #       5 
    neighbor_indices = indices[0][1:]   #   (0~N)
    neighbor_dists = distances[0][1:]   #  
    
    #    DataFrame  
    rec_real_indices = df_scaled.index[neighbor_indices]
    
    # --- Step 3:   ---
    print(f"\n Top 5 Recommendations for '{df_final.loc[target_idx, 'name']}':")
    rec_df = df_final.loc[rec_real_indices][['name', 'artists', 'year', 'popularity']].copy()
    rec_df['Distance'] = neighbor_dists #  0  
    print(rec_df)

    # --- Step 4:  (Deliverable 2: Input vs Neighbors) ---
    plt.figure(figsize=(12, 8))
    
    #   ()
    sns.scatterplot(x='PC1', y='PC2', data=pca_df, color='lightgray', alpha=0.2, s=10, label='All Songs')
    
    #   ( )
    rec_pca = pca_df.loc[rec_real_indices]
    plt.scatter(rec_pca['PC1'], rec_pca['PC2'], c='blue', s=150, label='Recommended')
    
    #   ( )
    input_pca = pca_df.loc[target_idx]
    plt.scatter(input_pca['PC1'], input_pca['PC2'], c='red', marker='*', s=400, zorder=10, label='Input Song')
    
    #    (    )
    plt.text(input_pca['PC1']+0.2, input_pca['PC2'], input_pca['name'], fontsize=12, fontweight='bold')
    for idx, row in rec_pca.iterrows():
        plt.text(row['PC1']+0.1, row['PC2'], row['name'], fontsize=10, color='darkblue')

    plt.title(f"Recommendation Map for '{df_final.loc[target_idx, 'name']}'")
    plt.legend()
    plt.grid(True, linestyle=':', alpha=0.5)
    plt.show()

# ==========================================
# 5. Execution Test
# ==========================================
# : Shape of You 
recommend_song("Shape of You")

#   (    )
# recommend_song("Bohemian Rhapsody")
# recommend_song("Idol", "BTS")
```
